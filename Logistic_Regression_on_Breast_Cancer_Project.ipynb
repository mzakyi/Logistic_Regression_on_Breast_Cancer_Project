# Cleaning Our Data

```python
# Before we clean our data, lets visualize to see if we have any null values.
sns.heatmap(data.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
# Since our target value would be diagnosis, lets convert it to 1s and 0s
data.diagnosis =[1 if value == "M" else 0 for value in data.diagnosis]
```

```python
#We can also plot the colum to see what it looks like using Seaborn
sns.barplot(x=data.diagnosis.value_counts().index, y=data.diagnosis.value_counts().values)
plt.ylabel("diagnosis")
plt.xlabel("counts")
plt.title("Class Distribution (0: Benign | 1: Malignant)")
```

# Logistic Regression

## Preprocessing

```python
# Now lets prepare our model
# y(diagnosis) is our target variable
# X is our predictors
y = data["diagnosis"]
X = data.drop(["diagnosis"], axis=1)
```

## Normalize the data

```python
# Since our predictors have different units we will have to normalize it first because splitting our data
from sklearn.preprocessing import StandardScaler

#Initializing the instance
normalizing_X = StandardScaler()

# Fit and transform the normalizing X data
X_normalized = normalizing_X.fit_transform(X)

# Now X is normalized also called X_normalized now, we can split the data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.30, random_state=100)
```

```python
#Now we call on the model we would like to build
from sklearn.linear_model import LogisticRegression

# Initilizing our logistic regression model
lr = LogisticRegression()

# Training the model on our training data
lr.fit(X_train, y_train)

# Predicting the target variable on the test data
y_predicted_Values = lr.predict(X_test)
```

```python
# We can see our predicted values are what we expected (1s and 0s)
y_predicted_Values
```

# Evaluating our mode

```python
from sklearn.metrics import accuracy_score

# Evaluating the accuracy of the model
accuracy = accuracy_score(y_test, y_predicted_Values)
print(f'Accuracy: {accuracy:.2f}')

#Our model is giving us a high score of 0.96 which is a pretty good indication that the model is efficient
```

```python
# We can also use other metrics from the sklearn such as classification report to measure the precisions our outputs
from sklearn.metrics import classification_report

print(classification_report(y_test,y_predicted_Values))
```
